```python
import os
import cv2
import numpy as np
import argparse
import matplotlib.pyplot as plt
from utils import plot_image, load_image, save_mask
```

```python
def get_number_of_closed_contours(contours):
    closed_contours = []
    for cnt in contours:
       if cv2.isContourConvex(cnt) == True:
          closed_contours.append(cnt)
       else:
          pass
    return len(closed_contours)
```

```python
def find_laplacian_of_gaussian(image):
    blur = cv2.GaussianBlur(image,(3,3),0)
    laplacian = cv2.Laplacian(image, cv2.CV_16S, ksize=3)
    result = cv2.convertScaleAbs(laplacian, alpha=2)
#     print(result.min(), result.max())
    return result
    
    
```

```python
def find_second_largest_contour(thresh):
#     plot_image(thresh)
    contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#     print(get_number_of_closed_contours(contours))
    contoursSorted = sorted(contours, key=lambda x: cv2.contourArea(x))
    mask = np.zeros(thresh.shape, np.uint8)
    cv2.drawContours(mask, contoursSorted, len(contoursSorted)-2, 255, cv2.FILLED)
#     plot_image(mask)
#     print(cv2.contourArea(contoursSorted[-2]))
    return mask

def segment_image(image, raw_image, contrast_factors=[1.5, 0]):
#     image = cv2.convertScaleAbs(image, alpha=1.5, beta=-10)
#     image = cv2.GaussianBlur(image,(3,3),0)
# plot_image(image)
    


    ret, thresh = cv2.threshold(raw_image, 50, 200, 0)
#     kernel = np.ones((7,7),np.uint8)
#     thresh = cv2.dilate(thresh, kernel, iterations = 1)

#     thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
#     thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    
#     plot_image(thresh)
#     thresh = np.bitwise_and(thresh, raw_image)
    thresh = np.bitwise_and(thresh, raw_image)
#     thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
#     thresh = cv2.dilate(thresh, kernel, iterations = 2)
#     thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
#     plot_image(thresh)
#     thresh = cv2.convertScaleAbs(thresh, alpha=1.5, beta=0)
#     plot_image(thresh)
    ret, thresh = cv2.threshold(thresh, 50, 200, 0)
#     plot_image(thresh)
#     thresh = np.bitwise_and(thresh, image)
#     ret, thresh = cv2.threshold(thresh, 50, 200, 0)
#     thresh = cv2.dilate(thresh, kernel, iterations = 3)
#     plot_image(thresh)
#     thresh = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
#     thresh = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 0)
#     plot_image(thresh)
#     kernel = np.ones((5,5),np.uint8)
#     thresh = cv2.dilate(thresh, kernel, iterations = 2)


    kernel = np.ones((3, 3),np.uint8)
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    
#     plot_image(thresh)
    result = find_second_largest_contour(thresh)
    
    while((result > 0).sum() < 20000):
        kernel = np.ones((3, 3), np.uint8)
        thresh = cv2.dilate(thresh, kernel, iterations = 1)
        result = find_second_largest_contour(thresh)
        
    return result
```
